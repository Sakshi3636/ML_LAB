```python
import re

# Define the regex pattern
regex_pattern = r'^(\d{4})'

# Test string
test_string = "2024 is the year of the next leap year."

# Search for the pattern in the test string
match = re.search(regex_pattern, test_string)

# Check if a match is found
if match:
    # Print the matched substring
    print("Match found:", match.group(0))
else:
    print("No match found.")
```
